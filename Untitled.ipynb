```python
import numpy as np
import scipy as scp
from scipy import integrate
import matplotlib.pyplot as plt

arrayOfValues=np.loadtxt(open("droptower_vdata.txt"))
arrayOfValues
```

```python
y=arrayOfValues[:,1]
x=arrayOfValues[:,0]
plt.plot(x,y)
distance=integrate.cumtrapz(y,x,-1,initial=0)
plt.plot(x,distance)
acceleration=np.diff(y)
plt.plot(x[0:10],acceleration)
```

```python
positive_acceleration=acceleration[acceleration>0]
avg=np.mean(positive_acceleration)
avg
```
